Make a model using Tanh function

```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import StandardScaler , LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# read data from local file
df = pd.read_csv(r"C:\Users\Lenovo\Downloads\covid_toy.csv")
df 
df["fever"].fillna(df["fever"].mean(),inplace = True)
```

```python
le = LabelEncoder()
df["gender"] = le.fit_transform(df["gender"])
df["cough"] = le.fit_transform(df["cough"])
df["city"] = le.fit_transform(df["city"])
df["has_covid"] = le.fit_transform(df["has_covid"])
```

```python
# make a target columns
X = df.drop("has_covid", axis=1)
y = df["has_covid"]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y ,test_size=.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
# make a model using a tanh function 
model = Sequential()
model.add(Dense(32 , input_dim = X_train.shape[1], activation="tanh"))
model.add(Dense(16 , activation="tanh"))
model.add(Dense(1 ,activation="sigmoid"))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fitting a model
model.fit(X_train , y_train , epochs=30 , batch_size=9 , verbose=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#  Evaluate  model
accuracy = model.evaluate(X_train, y_train)
print(f"Model Accuracy with Tanh: {accuracy:}")
```

```python
#   Predict and check output for first few test cases
y_pred = model.predict(X_test[:5])
print("Predicted probabilities (using Tanh activation in hidden layers):")
print(y_pred)
```
